### Model fitting and prediction
Let's fit a random forest regression model to the data and use it to make median house price predicions.
Use the default parameters, which includes using 100 base estimators, or regression trees.


````python
# Initialize and fit the Random Forest Regressor
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
rf_regressor.fit(X_train, y_train)

# Predict on test set
y_pred_test = rf_regressor.predict(X_test)
````

### Estimate out-of-sample MAE, MSE, RMSE, and R²


````python
mae = mean_absolute_error(y_test, y_pred_test)
mse = mean_squared_error(y_test, y_pred_test)
rmse = root_mean_squared_error(y_test, y_pred_test)
r2 = r2_score(y_test, y_pred_test)
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R² Score: {r2:.4f}")
````

### Interpretation of the Residual Plot

From the residual plot, we can infer a potential pattern indicating **heteroscedasticity** — the residuals are not evenly spread across all levels of the predicted house values. Specifically, it appears that:

- The residuals are **closer to zero** for houses with **median price values**.
- For **very low** or **very high house values**, the model tends to make **larger errors**, possibly underestimating expensive homes and overestimating cheaper ones.
  
This suggests that the model may not be capturing the full complexity of the data for extreme values, and a more flexible model (or feature engineering) might improve performance. However, for mid-range predictions, the model performs quite well.

Overall, the residuals do not appear to be completely random, which may hint at model bias or missing important predictive features.

### Exercise 8: Some Final Thoughts to Consider

- **Will the skewness affect the performance of Random Forest regression?**  
  Random Forest models are generally robust to skewed distributions because they are non-parametric and based on decision trees, which do not assume a normal distribution. However, extreme skewness may still affect performance indirectly by limiting the model’s ability to generalize well to outliers or underrepresented values.

- **Does the clipping of median house prices above $500,000 bias the predictions?**  
  Yes. If house values above $500,000 are clipped (capped), the model is unable to learn the true relationship between predictors and these higher values. This truncation introduces a ceiling effect, leading to systematic underprediction for expensive homes and a biased model.

- **Do we need to standardize the data?**  
  No, not for Random Forest. Unlike linear models or distance-based algorithms (e.g., SVM, k-NN), Random Forest does not rely on the scale of features because it splits data based on thresholds. Standardization is unnecessary and has little to no impact on Random Forest performance.
