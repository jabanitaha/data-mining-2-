```python
import pandas as pd

# Creating a dataset with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, None, 40, 28],  # Missing Age for Charlie
    'Salary': [50000, 60000, 55000, None, 58000],  # Missing Salary for David
    'City': ['New York', 'Los Angeles', None, 'Chicago', 'Houston']  # Missing City for Charlie
}

df = pd.DataFrame(data)
print("Original Dataset:\n")
df
```

```python
# **Calculating Means for Numeric Columns with Missing Values**
mean_age = df['Age'].mean()
mean_salary = df['Salary'].mean()

print("\nMean Age:", mean_age)
print("Mean Salary:", mean_salary)
```

```python
# Filling missing values with Mean (For Numeric Columns)
df_mean = df.copy()
df_mean = df_mean.assign(
    Age=df_mean['Age'].fillna(df_mean['Age'].mean()),
    Salary=df_mean['Salary'].fillna(df_mean['Salary'].mean()),
    City=df_mean['City'].fillna(df_mean['City'].mode()[0])  # Mode for categorical column
)

print("\nDataset after Mean Imputation:\n")
df_mean
```

df_mean = df.copy(): This line creates a copy of the original DataFrame df and assigns it to a new variable df_mean. This ensures that the original DataFrame remains unchanged.

df_mean = df_mean.assign(...): This is where the magic happens. The assign method is used to create new columns or modify existing ones. Inside assign

```python
import pandas as pd

# Creating a dataset with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, None, 40, 28],  # Missing Age for Charlie
    'Salary': [50000, 60000, 55000, None, 58000],  # Missing Salary for David
    'City': ['New York', 'Los Angeles', None, 'Chicago', 'Houston']  # Missing City for Charlie
}

df = pd.DataFrame(data)
print("Original Dataset:\n")
df
```

```python
# **Calculating Medians for Numeric Columns with Missing Values**
median_age = df['Age'].median()
median_salary = df['Salary'].median()

print("\nMedian Age:", median_age)
print("Median Salary:", median_salary)
```

```python
# Creating a copy of the dataset for Median Imputation
df_median = df.copy()

# **Applying Median Imputation**
df_median['Age'] = df_median['Age'].fillna(df_median['Age'].median())
df_median['Salary'] = df_median['Salary'].fillna(df_median['Salary'].median())

# **Applying Mode Imputation for Categorical Column (City)**
df_median['City'] = df_median['City'].fillna(df_median['City'].mode()[0])

print("\nDataset after Median and Mode Imputation:\n")
df_median
```

```python
import pandas as pd

# Creating a dataset with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, None, 40, 28],  # Missing Age for Charlie
    'Salary': [50000, 60000, 55000, None, 58000],  # Missing Salary for David
    'City': ['New York', 'Los Angeles', None, 'Chicago', 'Houston']  # Missing City for Charlie
}

df = pd.DataFrame(data)
print("Original Dataset:\n")
df
```

```python
# **5. Filling Missing Values with Mode**
df_mode = df.copy()

# Assigning mode values explicitly
df_mode['Age'] = df_mode['Age'].fillna(df_mode['Age'].mode()[0])
df_mode['Salary'] = df_mode['Salary'].fillna(df_mode['Salary'].mode()[0])
df_mode['City'] = df_mode['City'].fillna(df_mode['City'].mode()[0])  # Mode for City

print("\nDataset after Mode Imputation:\n")
df_mode
```

This line of code is specifically designed to fill missing values (represented as None or NaN) in the 'Age' column of the df_mode DataFrame using the mode.

Here's a step-by-step explanation:

df_mode['Age']: This selects the 'Age' column of the df_mode DataFrame. It isolates the column where we want to fill the missing values.

.fillna(...): This is a Pandas function applied to a DataFrame or Series to fill missing values. It takes an argument that specifies the value to use for filling.

df_mode['Age'].mode(): This part calculates the mode of the 'Age' column. The mode is the most frequent value in the column.

[0]: Since the .mode() function can return multiple values (if there are multiple modes), [0] is used to select the first mode value in the list. This ensures that a single value is used for filling.
Putting it Together:

df_mode['Age'].fillna(df_mode['Age'].mode()[0]) calculates the mode of the 'Age' column, takes the first mode value, and then uses that value to fill any missing values in the 'Age' column of the df_mode DataFrame.
df_mode['Age'] = ...: The result of the entire right-hand side expression (the filled 'Age' column) is then assigned back to the original df_mode['Age'] column. This effectively updates the DataFrame with the filled values


```python
import pandas as pd

# Creating a dataset with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, None, 40, 28],  # Missing Age for Charlie
    'Salary': [50000, 60000, 55000, None, 58000],  # Missing Salary for David
    'City': ['New York', 'Los Angeles', None, 'Chicago', 'Houston']  # Missing City for Charlie
}

df = pd.DataFrame(data)
print("Original Dataset:\n")
df
```

```python
# **Filling Missing Values with Mode for selected columns**
columns_to_impute = ['Age', 'Salary', 'City']

# Use a loop to fill missing values with mode for each column
for col in columns_to_impute:
    df[col] = df[col].fillna(df[col].mode()[0])

print("\nDataset after Mode Imputation:\n")
df
```

Data Normalization and Data Standardization

**Normalization scales the data into a specific range, typically [0, 1] or [-1, 1].** This is useful when the features have different units or magnitudes.

**Standardization transforms the data to have a mean of 0 and a standard deviation of **1, which is useful when the features have different scales.

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Generate sample data (10 data points, 2 features)
data = np.array([[10, 200],
                 [20, 250],
                 [30, 300],
                 [40, 350],
                 [50, 400],
                 [60, 450],
                 [70, 500],
                 [80, 550],
                 [90, 600],
                 [100, 650]])

# Convert to DataFrame for better visualization
df = pd.DataFrame(data, columns=['Feature1', 'Feature2'])

# Display original data
print("Original Data:")
df
```

import MinMaxScaler, StandardScaler: This is the core of the import statement. We're bringing in two specific classes:
**MinMaxScale**r: This is used for normalization, which scales your data to a specific range, typically between 0 and 1. It ensures that all your features have similar ranges of values, preventing features with larger values from dominating the model's learning process.
**StandardScaler**: This is used for standardization, which transforms your data to have a mean of 0 and a standard deviation of 1. It's useful when the features have different units or scales, and you want to ensure that they have equal influence on the model

```python
# --- Normalization (Min-Max Scaling) ---
scaler_minmax = MinMaxScaler()
normalized_data = scaler_minmax.fit_transform(df)

# Convert normalized data to DataFrame for better visualization
df_normalized = pd.DataFrame(normalized_data, columns=['Feature1', 'Feature2'])

# Display normalized data
print("\nNormalized Data (Min-Max Scaling):")
df_normalized
```

```python
# --- Standardization (Z-score Scaling) ---
scaler_standard = StandardScaler()
standardized_data = scaler_standard.fit_transform(df)

# Convert standardized data to DataFrame for better visualization
df_standardized = pd.DataFrame(standardized_data, columns=['Feature1', 'Feature2'])

# Display standardized data
print("\nStandardized Data (Z-score Scaling):")
df_standardized
```

Detecting Outliers Graphically

Outliers in data can often be detected visually using various types of plots. The most common graphical methods include:

Boxplot: A boxplot (also known as a whisker plot) visually shows the distribution of data, including the presence of outliers. Outliers are typically shown as points outside the "whiskers" of the boxplot.

Scatter Plot: In a scatter plot, outliers will appear as data points far away from the cluster of other points.

Histogram: A histogram can reveal outliers as bins that are distant from the main distribution

![image.png]()


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data with outliers
data = np.array([10, 12, 12, 13, 14, 15, 16, 17, 18, 100])

# Create a boxplot
figure= plt.figure(figsize=(5, 3))
plt.boxplot(data)
plt.title("Boxplot for Outlier Detection")
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy import stats

# Generate sample data with outliers
data = np.array([10, 12, 12, 13, 14, 15, 16, 17, 18, 100])

# Calculate the Z-scores
z_scores = stats.zscore(data)

# Define the threshold for identifying outliers (commonly 2.5 or 3)
threshold = 2.5  # Adjusted threshold to catch more subtle outliers

# Identify outliers
outliers = np.where(np.abs(z_scores) > threshold)

print("Z-scores:", z_scores)
print("")
print("Outliers indices:", outliers)
print("")
print("Outlier values:", data[outliers])
```
